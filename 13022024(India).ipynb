```python
import matplotlib.pyplot as plt

# Data
classes = ['6th', '7th', '8th', '9th', '10th', '11th', '12th']
num_students = [130, 120, 135, 130, 150, 80, 75]

# Plotting the bar chart
plt.figure(figsize=(10, 6))
plt.bar(classes, num_students, color='skyblue')

# Adding titles and labels
plt.title('Number of Students in Different Classes')
plt.xlabel('Classes')
plt.ylabel('Number of Students')

# Display the plot
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Plotting the pie chart
plt.figure(figsize=(8, 8))
plt.pie(num_students, labels=classes, autopct='%1.1f%%', startangle=140, colors=plt.cm.tab20.colors)

# Adding titles and labels
plt.title('Proportion of Students in Different Classes')

# Equal aspect ratio ensures that pie is drawn as a circle
plt.axis('equal')

# Display the plot
plt.tight_layout()
plt.show()

```
